## Scraping data from Olx in Delhi

```python
#let us connect to the webdriver
driver = webdriver.Chrome(r"C:/Users/atoranga/Documents/chromedriver.exe")
driver.get('https://www.olx.in/delhi_g4058659/cars_c84')
```

```python
brand= []
model = []
variant = []
YOM = []
kms = []
Fuel = []
owners = []
transmission = []
Location = []
Price =[]
```

```python
#Brand, model, variant, manufacturing year, driven kilometers, fuel, number of owners, location, Price 
links = []

#scrolling down to the webpage to load all the car details
for i in range(126):
    try:
        btn = driver.find_element_by_xpath('//button[@class="rui-39-wj rui-3evoE rui-1JPTg"]')
        btn.click()
        time.sleep(1)
        i+=1
    except NoSuchElementException as e:
        pass 
    
#collect the links from each car detail 
tag = driver.find_elements_by_xpath('//li[@class="_31j8e"]/a')

      
#fetching individual links of each car and fetch required details
for i in tag:
    links.append(i.get_attribute('href'))
    
```

```python
#first extract all tags where we have brand names
for i in links:
    driver.get(i) 
    try:
        #Scraping brand name
        name = driver.find_element_by_xpath('//div[@class="_35xN1"]')
        brand.append(name.text.split(' ')[0])
        
    except NoSuchElementException as e:
        pass

    try:  
        #scraping model name
        mdl = driver.find_element_by_xpath('//div[@class="_35xN1"]')
        model.append(mdl.text.split(' ')[1])
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping year of manufacturing      
        year = driver.find_element_by_xpath('//div[@class="_35xN1"]')
        YOM.append(year.text.replace(')','').split('(')[1])
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping total kms driven
        km = driver.find_element_by_xpath('//div[2][@class="_1u27F"]')
        kms.append(km.text.split(' ')[0])
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping fuel type of car        
        ful = driver.find_element_by_xpath('//div[1][@class="_1u27F"]')
        Fuel.append(ful.text)
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping transmission type of car
        trans = driver.find_element_by_xpath('//div[3][@class="_1u27F"]')
        transmission.append(trans.text)
    except NoSuchElementException as e:
        pass
    
    try:     
        #scraping car variant     
        var = driver.find_element_by_xpath('//div[@class="_3tLee"]')
        variant.append(var.text)
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping number of owners        
        ownr = driver.find_element_by_xpath('//div[1]/div[2]/div[2][@class="_1gasz"]')
        owners.append(ownr.text)
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping location of car        
        loc = driver.find_element_by_xpath('//*[@id="container"]/main/div/div/div/div[1]/div/ol/li[4]/a')
        Location.append(loc.text.split(' ')[2])
    except NoSuchElementException as e:
        pass
    
    try:        
        #scraping price of car        
        pric = driver.find_element_by_xpath('//div[@class="_3FkyT"]')
        Price.append(pric.text.replace('₹',''))        
    except NoSuchElementException as e:
        pass
    
    finally:
        pass
    
    
col =[brand,model,variant,YOM,kms,Fuel,owners,transmission,Location,Price]
for i in col:
    print(len(i), end=' ')
olx_delhi = pd.DataFrame({'Brand':brand, 'Model':model, 'Variant':variant,'Year of Manufacture': YOM, 'Driven Kilometers':kms, 'Fuel type':Fuel,'Transmission Type':transmission, 'Number of Owners': owners,'Location':Location,'Price in Lakhs':Price})
olx_delhi
```

```python
#let us connect to the webdriver
driver = webdriver.Chrome(r"C:/Users/atoranga/Documents/chromedriver.exe")
driver.get('https://www.carwale.com/used/cars-for-sale/#sc=-1&so=-1&city=3001&pc=3001')
time.sleep(2)

#empty lists
brand= []
model = []
variant = []
YOM = []
kms = []
Fuel = []
owners = []
transmission = []
Location = []
Price =[]


#loading more number of car details
SCROLL_PAUSE_TIME = 0.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
    
time.sleep(2)
#Brand, model, variant, manufacturing year, driven kilometers, fuel, number of owners, location, Price 
links = []
 
#collect the links from each car detail 
tag = driver.find_elements_by_xpath('//h2[@class="card-detail-block__title"]/a')

#fetching individual links of each car and fetch required details
for i in tag:
    links.append(i.get_attribute('href'))
    
    try:
        #scraping location of car        
        loc = driver.find_element_by_xpath('//span[@class="cityName vehicle-data__item"]')
        Location.append(loc.text)
         
    except NoSuchElementException as e:
        pass
                
time.sleep(2)
#first extract all tags where we have brand names
for i in links:
    driver.get(i) 
    try:
        #Scraping brand name
        name = driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[4]/a/span")
        brand.append(name.text)
             
    except NoSuchElementException as e:
        pass
    
    try:  
        #scraping model name
        mdl = driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[5]/a/span")
        model.append(mdl.text)
        
    except NoSuchElementException as e:
        pass
    
    try:     
        #scraping car variant     
        var = driver.find_element_by_xpath('//span[@class="font16 text-light-grey text-unbold margin-left10 special-skin-text"]')
        variant.append(var.text)
         
    except NoSuchElementException as e:
        pass
        
    try:
        #scraping year of manufacturing      
        year = driver.find_element_by_xpath("//ul[@class='basic-details']/li[1]/span[2]")
        YOM.append(year.text)
         
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping total kms driven
        km = driver.find_element_by_xpath("//ul[@class='basic-details']/li[2]/span[2]")
        kms.append(km.text)
         
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping fuel type of car        
        ful = driver.find_element_by_xpath("//ul[@class='basic-details']/li[3]/span[2]")
        Fuel.append(ful.text)
        
    except NoSuchElementException as e:
        pass
    
    try:
        #scraping transmission type of car
        trans = driver.find_element_by_xpath("//ul[@class='basic-details']/li[4]/span[2]")
        transmission.append(trans.text)
         
    except NoSuchElementException as e:
        pass
     
    try:
        #scraping number of owners        
        ownr = driver.find_element_by_xpath('//*[@id="overview"]/div/ul/li[9]/div[2]')
        owners.append(ownr.text)
         
    except NoSuchElementException as e:
        pass
    
       
    try:        
        #scraping price of car        
        pric = driver.find_element_by_xpath('/html/body/div[14]/section[2]/div[2]/div[2]/div/div[1]/div[1]/div/div[1]/p')
        Price.append(pric.text.replace('₹',' ').split(' ')[2]) 
        
        
    except NoSuchElementException as e:
        pass
    
    finally:
        pass
    
    
col =[brand,model,variant,YOM,kms,Fuel,owners,transmission,Location,Price]
for i in col:
    print(len(i), end=' ')
carwale_delhi = pd.DataFrame({'Brand':brand, 'Model':model, 'Variant':variant,'Year of Manufacture': YOM, 'Driven Kilometers':kms, 'Fuel type':Fuel,'Transmission Type':transmission, 'Number of Owners': owners,'Location':Location,'Price in Lakhs':Price})
carwale_delhi
    
    
```
